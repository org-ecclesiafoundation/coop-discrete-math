## Properties of Addition

### Associative Property

The way numbers are grouped in an operation (addition or multiplication) does not change the result. It applies to both addition and multiplication, but not to subtraction and division.

$(a+b)+c = a+(b+c)$

Think of standing in a line from left to right in the order: 

#### Me | You | Mobster with a Bad Attitude

If you shake hands with me first, he'll shoot you; if you shake hands with him first, you're fine. That would mean that, because he's on your right. Shaking hands is a right-associative property for you, because you need to operate with the thing on the right first.

Now, think of standing in line the other way:

#### Mobster with a Bad Attitude | You | Me

If the angry mobster is on your left, and you need to shake hands with him first to keep from getting shot, then that would mean that shaking hands is left-associative for you.

#### For Subtraction, Division, and Combining Unlike Operations

- Subtraction and division are left-associative. You apply the operations from left to right instead of right to left.

- If you are combining multiplication and division, the signs are left-associative.

- If you are combining addition and subtraction, the signs are left-associative.

- If you are combining something other than the stuff above that, you follow PEMDAS first, and then apply associativity.

Don't ask why this is. The reason is that if we didn't settle on something, everyone would be arguing about how to do simple arithmetic and we can't have that happen and expect to actually function in the world without going to war over minus-signs.

### Commutative Property

Latin word "commutare" means "to change". That means that if you change the position of two operands, the result is the same

$a+b=b+a$

### Distributive Property

This property is used to multiply or divide a single term and two or more terms inside a set of parentheses. It provides a way to multiply numbers by breaking them down into smaller parts.

$a\times(b+c) = a\times b + a\times c$

$a\div(b+c) = a\div b + a\div c$

### Identity Property

There is an operation you can apply to numbers that leaves them unchanged: adding zero, subtracting zero, multiplying by 1, and dividing by 1.

$a+0=a$

$a-0=a$

$a\times1=a$

$a\div1=a$

## Decimals and Fractions

What helps me is separating the fractions from the non-fractions.

$10.8 = 10 + 0.8$

$42\frac{4}{5} = 42 + \frac{4}{5}$

Now, what you do is deal with the non-fractions, deal with the fractions, and then combine the result:

$10 + 42 = 52$

$0.8 + \frac{4}{5} = ???$

You have two choices: You can either write things in terms of decimals, or write things in terms of fractions. You probably learned some conversion thingo in class, but I'll just show you how I do it:

$0.8 = \frac{8}{10}$

### The Naive Way

The naive way is to multiply $\frac{8}{10}\times\frac{5}{5}=\frac{40}{50}$ and $\frac{4}{5}\times\frac{10}{10}=\frac{40}{50}$

Add the fractions: $\frac{40}{50}+\frac{40}{50}=\frac{80}{50}$, which is the same as $1\frac{30}{50}$

Reduce it to lowest terms and add to the 52 you calculated from above:

$52+1\frac{4}{5}=53\frac{3}{5}$

QED

### The Quick Way

The quick way is to know that if you divide 1 into fifths, you get

$\frac{1}{5}=0.2$

$\frac{2}{5}=0.4$

$\frac{3}{5}=0.6$

$\frac{4}{5}=0.8$

$\frac{5}{5}=1.0$

Then, you do the math in fifths and don't worry about having to reduce to lowest terms at the end.

## Increases and Decreases

If something increases by an amount, you add; if something decreases by an amount you subtract. Don't overthink it.

If you cut off my feet and I get shorter (my height decreases) by 4 inches, my height is $66in - 4in = 62in$

If I put on platform shoes and I get taller (my height increases) by 4 inches, my height is $66in + 4in = 70in$

#### Problem 9

$-3.75-(-6\frac{1}{4})$

Get rid of the double negatives first:

$-3.75+6\frac{1}{4}$

The quick way is to think about quarters being $0.25. $0.75 is three quarters of a dollar, so we have:

$-3\frac{3}{4}+6\frac{1}{4}$

It might be easier to reverse terms; it might not. That's totally up to you

$6\frac{1}{4}-3\frac{3}{4}=2\frac{1}{2}$

#### Problem 10

$-2\frac{1}{4}+5\frac{3}{4}+2.25$
